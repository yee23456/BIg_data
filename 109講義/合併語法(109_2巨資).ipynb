```
import pandas as pd
import numpy as np
```

# 一、merge合併模式

```
#輸入兩行資料key和data1
df1 = pd.DataFrame({'key':['c','b','a','c','a','a','b'],'data1': range(7)})
df1
```

```
df2 = pd.DataFrame({'key':['a','b','d'],'data2':range(3)})
df2
```

```
#inner交集合併模式, on:'key'依此欄位合併
df1.merge(df2,on = 'key',how = 'inner')#內連線，取交集
```

```
#outer聯集合併模式
df1.merge(df2,on = 'key',how = 'outer')#外連結，取連集，並用nan填充
```

```
#left左側檔案為主之合併模式
df1.merge(df2,on = 'key',how = 'left')#左連線，左側DataFrame取全部，右側DataFrame取部分
```

```
#right右側檔案為主之合併模式
df1.merge(df2,on = 'key',how = 'right')#右連線，右側DataFrame取全部，左側DataFrame取部分
```

# 二、pd.merge模式

```
#對連線表中非連線列的重複列名的處理
left = pd.DataFrame({'key1':['foo','foo','bar'],'key2':['one','one','two'],'lval':[1,2,3]})
right = pd.DataFrame({'key1':['foo','foo','bar','bar'],'key2':['one','one','one','two'],'rval':[4,5,6,7]})
```

```
left
```

```
right
```

```
pd.merge(left,right,on = 'key1')
```

```
#若合併的欄位名稱不相同時，可採此語法
pd.merge(left,right,left_on=('key1'), right_on=('key1'))
```

```
#若合併的欄位名稱不相同時，亦可採此語法
left.merge(right, left_on=('key1'), right_on=('key1'))
```

# 三、concat合併模式

```
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'], index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['b','c','d','e'], index=[2,3,4])
```

```
df1
```

```
df2
```

```
#concat:連結;axis-0直行合併,取onter行取聯集
res = pd.concat([df1, df2], axis=0, join='outer')
res
```

```
#axis=0直行合併，取inner行取交集
res = pd.concat([df1, df2], axis=0, join='inner')
res
```

```
#重置index, 0,1,2,...
res = pd.concat([df1, df2], axis=0, ignore_index=True)
res
```

```
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'], index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['b','c','d','e'], index=[2,3,4])
```

```
df1
```

```
df2
```

```
#axis=1橫向合併
res = pd.concat([df1, df2], axis=1)
print(res)
```
