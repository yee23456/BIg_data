# 相關分析:用於分析兩變數是否有顯著相關(量*量)

```
#利用numpy之相關性corr相關分析的語法
import numpy as np
b=np.corrcoef(df["3P"],df["PTS"])
b[1,0]
```

```
#利用scipy函式庫中的stats功能，計算皮爾森相關係數(pearsonr)及檢定分析之P值
from scipy import stats
(a,pv)=stats.pearsonr(df['PTS'],df["3P"]) 
print('r = %.2f,  p_value= %.2f'%(a,pv))
```

```
#同時想一次進行很多變數的相關分析
#量的資料
qu=['DIFF', '1Q',
       '2Q', '3Q', '4Q', 'F', 'MIN', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA',
       'OR', 'ORR', 'DRR', 'FT%', 'TOV%', 'EFG%', 'DRB%', 'ORB%', 'DR', 'TOT',
       'A', 'PF', 'ST', 'TO', 'BL', 'PTS', 'POSS', 'PACE', 'OEFF', 'DEFF',
       'REST DAYS', 'OPENING ODDS', 'SPREAD']
len(qu)
```

```
#利用scipy之相關性corr相關分析的語法
#皮爾森相關分析(量*量)
from scipy import stats
name=[]
alist=[]
pvlist=[]
for i in range(len(qu)):
  #取qu變數與PTS做皮爾森相關分析，a為相關係數，pv為P值
  (a,pv)=stats.pearsonr(df['PTS'],df[qu[i]]) 
  name.append(qu[i])
  alist.append(a)
  pvlist.append(pv)
print(alist)
```

```
#display.float_format:接受浮點數並具有格式的字串，bug修復顯示格式以避免運行時錯誤
#lambda:定義函式，執行運算式時將會產生函式物件
#取6位小數點
pd.set_option('display.float_format',lambda x:'%.5f'%x)
#資料結構以DataFrame顯示 
#語法:pd.DataFrame(資料,index名稱,columns名稱)
result=pd.DataFrame(alist,name,columns=['相關係數'])
#將pvlist加進result第二欄
result.insert(1,'p-value',pvlist) 
result
```

# t檢定:用於比較兩樣本的平均數是否有顯著差異(類*量)

```
#t檢定:用於比較兩樣本的平均數是否有顯著差異(類*量)
#檢測失敗與勝利其3P三分球進球數是否有差異?
#匯入scipy模組
from scipy import stats
#語法:stats.ttest_ind(樣本1x,樣本2y)
#取當W/L=0 (失敗)時的3P與W/L=1(勝利)時的3P去做t檢定
(t,p)=stats.ttest_ind(df['PTS'][df['W/L']==0],df['PTS'][df['W/L']==1])
print('t value= %.2f,  p_value= %.2f'%(t,p))
```

```
#列出失敗與勝利平均三分球之進球數數
a=df.groupby('W/L').mean()
b=a[["PTS"]]
b
```

```
#列出失敗與勝利平均三分球之進球數數
c=df.groupby('W/L').count()
d=c[["DATASET"]]
d
```

```
d.merge(b,on="W/L")
```

```
#同時想一次進行很多變數的t檢定
#把變數放進名為qv2的陣列中
qv2=['DIFF', '1Q',
       '2Q', '3Q', '4Q', 'F', 'MIN', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA',
       'OR', 'ORR', 'DRR', 'FT%', 'TOV%', 'EFG%', 'DRB%', 'ORB%', 'DR', 'TOT',
       'A', 'PF', 'ST', 'TO', 'BL', 'PTS', 'POSS', 'PACE', 'OEFF', 'DEFF',
       'REST DAYS', 'OPENING ODDS', 'SPREAD']
```

```
name=[]
alist=[]
pvlist=[]
for i in range(len(qv2)):
  #分別取t_test變數與W/L=0及W/L=1的交集去做t檢定
  (a,pv)=stats.ttest_ind(df[qv2[i]][df['W/L']==0],df[qv2[i]][df['W/L']==1])
  #將t_test項目加進name中
  name.append(qv2[i])
  #將a(t值)加進alist中
  alist.append(a)
  #將pv值加進pvlist中
  pvlist.append(pv)
```

```
#lambda表示式是一種效能比較好的迴圈,跑比較快
pd.set_option('display.float_format',lambda x:'%.2f'%x)
#將alist資料結構用DataFrame顯示出
result=pd.DataFrame(alist,name,columns=['t-value'])
#將pvlist加進result第二欄
result.insert(1,'p-value',pvlist)
result
```

```
#變異數分析:比較多組樣本之間的平均數差異(類*量)
#檢測不同的REST DAYS其3P三分球進球數是否有差異
from scipy import stats
(f,pv)=stats.f_oneway(df['PTS'][df['REST DAYS']==0],
                          df['PTS'][df['REST DAYS']==1],
                          df['PTS'][df['REST DAYS']==2],
                          df['PTS'][df['REST DAYS']==3])
print('%.3f,%.3f'%(f,pv))
```

```
d.merge(b,on="REST DAYS")
```

```
#同時想一次進行很多變數的單因子變異數分析oneway ANOVA
#把變數放進名為one的陣列中
one=['DIFF', '1Q',
       '2Q', '3Q', '4Q', 'F', 'MIN', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA',
       'OR', 'ORR', 'DRR', 'FT%', 'TOV%', 'EFG%', 'DRB%', 'ORB%', 'DR', 'TOT',
       'A', 'PF', 'ST', 'TO', 'BL', 'PTS', 'POSS', 'PACE', 'OEFF', 'DEFF',
       'REST DAYS', 'OPENING ODDS', 'SPREAD']
name2=[]
alist2=[]
pvlist2=[]
for i in range(len(one)):
  (a2,pv2)=stats.f_oneway(df[one[i]][df['REST DAYS']==0],
                          df[one[i]][df['REST DAYS']==1],
                          df[one[i]][df['REST DAYS']==2],
                          df[one[i]][df['REST DAYS']==3])
  name2.append(one[i])
  alist2.append(a2)
  pvlist2.append(pv2)

result2=pd.DataFrame(alist2,name2,columns=['F_value'])
result2.insert(1,'p-value',pvlist2)
result2
```

# 卡方檢定:想檢定兩個類別變數彼此間是否相互獨立(類VS類)

```
#交叉表:兩個變項間的分佈狀況
#休息天數(REST DAYS)VS 勝負(W/L)的分佈狀況
#語法:pd.crosstab(index,columns,margins=True)，margins=True表顯示all，False或不命令margins則不顯示
hh=pd.crosstab(df['W/L'],df['REST DAYS'])
hh
```

```
#卡方檢定:想檢定兩個類別變數彼此間是否相互獨立
#分析不同休息天數(REST DAYS)VS 勝負(W/L)的差異；亦即勝負與休息天數是否顯著關係)
#匯入scipy.stats模組
#卡方檢定依序顯示卡方值、P值、自由度、與原數據同維度的對應期望值
import scipy.stats as stats
cr=stats.chi2_contingency(hh)
print("Chi-Square value=%.3f,   p_value=%.3f"%cr[0:2])
```
