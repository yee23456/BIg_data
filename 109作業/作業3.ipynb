````python
#第五題 請查詢台灣近年來失業率與平均所得(或所得中位數)的資料，並劃出失業率與平均所得(或所得中位數)之趨勢圖。
df1 = pd.read_excel('平均所得.xls')
df2 = pd.read_csv('失業率_年.csv',encoding='big5')

x = df2.DATE
y1 = df1.iloc[7:50,8]/100000 #平均所得總額
y2 = df2.data         #失業率

plt.title(u'失業率與平均所得(或所得中位數，單位:萬)之趨勢圖',fontproperties=zhfont,fontsize=20)
plt.xlabel('Date',fontsize=20)
plt.ylabel(u'錢(單位:萬) & 對應比例值',fontproperties=zhfont,fontsize=20)
plt.plot(x, y1, 'b--o',label="per capita income")
plt.plot(x,y2,'r-o',label="Unemployment Rate")
plt.legend() 
````
